```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "http://lhr.nu.edu.pk/faculty/"
r = requests.get(URL)
print(r)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
#print(soup.prettify())
```

```python
soup
```

```python
dp=['fsc','ee','cv','fsm','ss']
```

```python
f= soup.find('div',{'id':'fsc'})
fac = f.findAll('div',class_='facultyCard')
faculty=fac[0]
```

```python
id = int(faculty.find('a',class_='faculty-link')['href'].split('/')[-1])
id
```

```python
h = 'https://lhr.nu.edu.pk/'+'fsc'+'/facultyProfile/'+'1238'
r1 = requests.get(h)
soup1 = BeautifulSoup(r1.content, 'html.parser')
f8=soup1.find('div', class_='facultyCard')
f9= soup1.find('span',class_='small').text.split(':')[1]
f9
```

```python
id_=[]
name=[]
email=[]
desig=[]
img=[]
hec=[]
highest=[]
ext=[]
dep=[]

val=0
for j in range(len(dp)):
   f= soup.find('div',{'id':dp[j]})
   fac = f.findAll('div',class_='facultyCard')
   print('Len:' + dp[j])
   print(len(fac))
   for i in range(len(fac)):
    faculty = fac[i]
    try:
        id_.append(int(faculty.find('a',class_='faculty-link')['href'].split('/')[-1]))
    except:
        id_.append(-1)
    try:
        hec.append(bool(faculty.find('p', class_='font-italic').text.split('\n')[2].strip().split(' ')[1]))
    except:
        hec.append(bool(0))
    name.append(faculty.find('h5', class_='text-center').text)
    desig.append(faculty.find('p',class_='small').text.strip().split('\n')[0])
    email.append(faculty.find('p', class_='mb-0').text)
    img.append('https://lhr.nu.edu.pk'+faculty.find('img', class_='card-img-top')['src'])
    faculty_link = 'https://lhr.nu.edu.pk/'+dp[j]+'/facultyProfile/'+str(id_[val])
    val=val+1
    dep.append(dp[j])
    r1 = requests.get(faculty_link)
    soup1 = BeautifulSoup(r1.content, 'html.parser')
    f8=soup1.findAll('div', class_='facultyCard')
    try:
        ext.append(soup1.find('span',class_='small').text.split(':')[1])
    except:
        ext.append(-1)
    try:
      highest.append(soup1.find('div', class_='text-justify').find('li').text.strip())
    except:
      highest.append('Not Available')


```

```python
data = {
    'id': id_,
    'name': name,
    'email': email,
    'desig': desig,
    'img': img,
    'hec': hec,
    'highest': highest,
    'ext': ext,
    'dep': dep
}

df = pd.DataFrame(data)
```
